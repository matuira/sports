## Анализ массива данных

```python
import numpy as np
import pandas as pd
import matplotlib
```

```python
df = pd.read_csv('dataset_Facebook.csv', sep=';')
```

### Статистика по всему массиву

```python
df
```

```python
df.describe()
```

Заметим, что столбцы Paid, like, chare содержат пропущенные значения (общее число записей меньше 500). Такие данные портят статистику. Следует либо удалить соответствующие записи(строки), либо заполнить их с помощью средних значений. Поскольку изначальных данных не так много, выберем второй способ. 

```python
df = df.fillna(df.median(axis=0), axis=0)
```

И еще раз посмотрим статискику по всему массиву данных, уже без пропущенных значений. 

```python
df.describe()
```

Видим, что есть объект с максимальным значением лайков - 5172, число которых значительно превосходит среднее значение - 178.

```python
df.loc[df['like']== 5172]
```

Это тот же объект, которому принадлежит рекорд по share, total interaction. Его и стоит назвать **самым популярным** объектом в выборке.
